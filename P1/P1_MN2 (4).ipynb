```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
L = 1.0           # Longitud normalizada de la barra
T_max = 0.025     # Tiempo final de inters
dx = 0.01         # Paso espacial solicitado
gamma = 0.49      # Parmetro de estabilidad
dt = gamma * dx**2 # Paso temporal segn gamma
N_x = int(L / dx) + 1 # Nmero de puntos espaciales
N_t = int(T_max / dt) + 1 # Nmero de pasos temporales

# Inicializa la matriz de temperaturas
T = np.zeros((N_x, N_t))

# Condicin inicial
T[:, 0] = 0  # Toda la barra comienza a 0C

# Condiciones de frontera
T[0, :] = 0  # Extremo izquierdo fijo
T[-1, :] = 0 # Extremo derecho fijo

# Itera en el tiempo
for n in range(0, N_t - 1):
    for i in range(1, N_x - 1):
        T[i, n+1] = T[i, n] + gamma * (T[i+1, n] - 2*T[i, n] + T[i-1, n]) + dt

# Extrae la solucin en t = 0.025
x = np.linspace(0, L, N_x)  # Puntos espaciales
t_target_index = int(T_max / dt)  # ndice correspondiente al tiempo t = 0.025

# Grfica
plt.plot(x, T[:, t_target_index], label=f"T(x, t={T_max})")
plt.xlabel("x (posicin)")
plt.ylabel("T (temperatura)")
plt.title(f"Distribucin de temperatura en t={T_max}")
plt.legend()
plt.grid()
plt.show()
```
